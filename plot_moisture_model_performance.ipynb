```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Final model coefficients
a = 0.00000140
b = 0.00046
c = 20.29594

# EC (S/cm) values from 3rd session
ec_values = np.array([
    775.75, 570.95, 287.75, 335.6, 180.55,
    872.45, 2679.9, 214.5,
    194.75, 703.65, 104.1, 310.15, 274.5,
    381.7, 662.75, 563.2, 1659.1, 60.75
])


# Lab moisture values (desired)
lab_moisture = np.array([
    19.7, 20.9, 23.0, 23.6, 21.0,
    23.7, 31.6, 21.6,
    18.9, 19.9, 18.0, 20.7, 20.6,
    19.8, 20.8, 20.5, 24.9, 20.0
])

# Calculate predicted moisture using 2nd-degree regression
predicted_moisture = (a * ec_values ** 2) + (b * ec_values) + c

# Calculate percent error
percent_error = np.abs(predicted_moisture - lab_moisture) / lab_moisture * 100

# Plot predicted vs actual
plt.figure(figsize=(12, 5))

# Subplot 1: Actual vs Predicted
plt.subplot(1, 2, 1)
plt.plot(lab_moisture, label='Lab Moisture %', marker='o')
plt.plot(predicted_moisture, label='Predicted Moisture %', marker='x')
plt.title('Predicted vs Lab Moisture')
plt.xlabel('Sample Number')
plt.ylabel('Moisture %')
plt.legend()
plt.grid(True)

# Subplot 2: Percent Error
plt.subplot(1, 2, 2)
plt.bar(range(1, len(percent_error)+1), percent_error, color='orange')
plt.axhline(10, color='red', linestyle='--', label='10% Threshold')
plt.title('Percent Error per Sample')
plt.xlabel('Sample Number')
plt.ylabel('Percent Error (%)')
plt.xticks(range(1, len(percent_error)+1))
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
